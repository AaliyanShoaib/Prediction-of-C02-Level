```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv('/content/global_co2.csv')
dataset
```

```python
dataset.info()
missingvalues = dataset.isnull().sum()
print ("Missing values:\n" , missingvalues)
```

```python
w=dataset.dropna()
w.info()
```

```python
s=pd.DataFrame(w)
s
```

```python
x= w.iloc[:,:1].values
y= w.iloc[:,-1].values
print (x)
k=pd.DataFrame(x)
k
```

```python
x = w.iloc[: , :1].values  #rows, columns
y = w.iloc[: , -1:].values
print (y)
```

**TRAINING TESTING AND SPLITTING**

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train, y_test = train_test_split (x,y,test_size=0.1,random_state=1)
```

```python
print ("Value of X train : \n" , X_train )
print ("Value of X test : \n" , X_test )
print ("Value of Y train : \n" , y_train )
print ("Value of Y test : \n" , y_test )
```

**LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression
lte= LinearRegression()
lte.fit(x,y)
```

```python
plt.scatter(X_train,y_train,color = 'red') #FOR TRAINING SET
plt.plot(X_train,lte.predict(X_train),color = 'blue')
plt.title(' Linear Regression ')
plt.xlabel('Years')
plt.ylabel('CO2')
plt.show()
print("Accuracy: ",lte.score(X_test,y_test)* 100)
```

```python
plt.scatter(X_test,y_test,color = 'red') #FOR TESTING SET
plt.plot(X_train,lte.predict(X_train),color = 'blue')
```

```python
print (lte.predict([[2011]]))
print (lte.predict([[2012]]))
print (lte.predict([[2013]]))
```

**POLYNOMIAL REGRESSION**

```python
from sklearn.preprocessing import PolynomialFeatures
poly_fit=PolynomialFeatures(degree=90)
k=poly_fit.fit_transform(X_train)
k_test=poly_fit.transform(X_test)

lte2=LinearRegression()
lte2.fit(k,y_train)
ypred= lte2.predict(k)
ypredtest= lte2.predict(k_test)
```

```python
x_grid=np.arange(min(X_train),max(X_train),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
x_grid_poly = poly_fit.transform(x_grid)
plt.scatter(X_train,y_train,color = 'red')
plt.plot(x_grid,lte2.predict(x_grid_poly),color='blue')
train_score = lte2.score(k,y_train)
test_score = lte2.score(k_test,y_test)
print("Training Accuracy  ", train_score * 100)

```

```python
x_grid=np.arange(min(X_test),max(X_test),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
x_grid_poly = poly_fit.transform(x_grid)
plt.scatter(X_test,y_test,color = 'red')
plt.plot(x_grid,lte2.predict(x_grid_poly),color='blue')
train_score = lte2.score(k,y_train)
test_score = lte2.score(k_test,y_test)
print("Training Accuracy  ", train_score * 100)
```

```python
print ('2011 Prediction : ' , lte2.predict(poly_fit.fit_transform([[2011]])))
print ('2012 Prediction : ' , lte2.predict(poly_fit.fit_transform([[2012]])))
print ('2013 Prediction : ' , lte2.predict(poly_fit.fit_transform([[2013]])))
```

**DECISION TREE REGRESSION**

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X_train,y_train)
train2_score = dtr.score(X_train,y_train)
test2_score = dtr.score(X_test,y_test)
```

```python
x_grid=np.arange(min(X_train),max(X_train),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(X_train,y_train,color='red')
plt.plot(x_grid,dtr.predict(x_grid),color='blue')
plt.show()
print("Training Accuracy  ", train2_score * 100)
```

```python
x_grid=np.arange(min(X_test),max(X_test),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(X_test,y_test,color='red')
plt.plot(x_grid,dtr.predict(x_grid),color='blue')
plt.show()

print("Testing Accuracy  ", test2_score * 100)
```

```python
print ('2011 Prediction : ' , dtr.predict([[2011]]))
print ('2012 Prediction : ' , dtr.predict([[2012]]))
print ('2013 Prediction : ' , dtr.predict([[2013]]))
```

**RANDOM FOREST REGRESSION**

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 50 , random_state= 0)
regressor.fit(X_train,y_train)
```

```python
x_grid=np.arange(min(X_train),max(X_train),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(X_train,y_train,color='red')
plt.plot(x_grid,regressor.predict(x_grid),color='blue')
plt.show()
tscore = regressor.score(X_train,y_train)
print("Training Accuracy  ", tscore * 100)
```

```python
x_grid=np.arange(min(X_test),max(X_test),0.01)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(X_test,y_test,color='red')
plt.plot(x_grid,regressor.predict(x_grid),color='blue')
plt.show()
testscore = regressor.score(X_test,y_test)
print("Testing Accuracy  ", testscore * 100)
```

```python
print ('2011 Prediction : ' , regressor.predict([[2011]]))
print ('2012 Prediction : ' , regressor.predict([[2012]]))
print ('2013 Prediction : ' , regressor.predict([[2013]]))
```
